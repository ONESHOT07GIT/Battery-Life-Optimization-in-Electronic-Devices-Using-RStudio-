### Markowitz's mean-variance **optimization**

```python
from scipy.optimize import minimize
mean_returns = simple_returns.mean()
cov_matrix = simple_returns.cov()
initial_weights = np.ones(len(tickers)) / len(tickers)

# objective function for minimization (negative of portfolio return)
def objective(weights):
    return -np.dot(weights, mean_returns)

#for the weights (sum of weights = 1)
constraints = ({'type': 'eq', 'fun': lambda weights: np.sum(weights) - 1})

# Define the bounds for the weights (any w such that 1<= w <=0)
bounds = tuple((0, 1) for asset in range(len(tickers)))

# Initialize arrays to store efficient frontier points
efficient_returns = []
efficient_volatilities = []
efficient_weights = []

# Iterate through different levels of expected return
for target_return in np.linspace(mean_returns.min(), mean_returns.max(), num=100):
    # Define the constraint for the target return
    constraints = ({'type': 'eq', 'fun': lambda weights: np.sum(weights) - 1},
                   {'type': 'eq', 'fun': lambda weights: np.dot(weights, mean_returns) - target_return})

    # Perform mean-variance optimization
    result = minimize(objective, initial_weights, method='SLSQP', bounds=bounds, constraints=constraints)

    # Extract optimal weights from the result
    optimal_weights = result.x

    # Calculate the expected return and volatility for the optimal portfolio
    efficient_return = -result.fun
    efficient_volatility = np.sqrt(np.dot(optimal_weights.T, np.dot(cov_matrix, optimal_weights)))

    # Append the values to the arrays
    efficient_returns.append(efficient_return)
    efficient_volatilities.append(efficient_volatility)
    efficient_weights.append(optimal_weights)

# Convert the lists to NumPy arrays
efficient_returns = np.array(efficient_returns)
efficient_volatilities = np.array(efficient_volatilities)


```

```python
# Print Optimal Portfolios on the Efficient Frontier
print("Optimal Portfolios on the Efficient Frontier:")
for i in range(len(efficient_returns)):
    print(f"Portfolio {i+1}:")
    print(f"  Return     : {efficient_returns[i]:.6f}")
    print(f"  Volatility : {efficient_volatilities[i]:.6f}")
    print("  Weights    : ", ", ".join([f"{tickers[j]}: {efficient_weights[i][j]:.6f}" for j in range(len(tickers))]))
    print()
```

```python
import pandas as pd

# Create a DataFrame for the optimal portfolios
portfolio_data = {
    "Portfolio": [f"Portfolio {i+1}" for i in range(len(efficient_returns))],
    "Return": efficient_returns,
    "Volatility": efficient_volatilities
}

# Add weights for each asset
for i in range(len(tickers)):
    portfolio_data[tickers[i]] = [efficient_weights[j][i] for j in range(len(efficient_returns))]

# Create DataFrame
portfolio_df = pd.DataFrame(portfolio_data)

# Save DataFrame as CSV
portfolio_df.to_csv("efficient_frontier_portfolios.csv", index=False)

# Display DataFrame
print(portfolio_df)
```

 **Risk measures used are beta, standard deviation, and the Sharpe ratio.**

```python
market_index = "VOO" ## assumption
risk_measures = pd.DataFrame(index=tickers, columns=['Standard Deviation', 'Beta', 'Sharpe Ratio'])

for ticker in tickers:
    # Calculate standard deviation
    std_dev = simple_returns[ticker].std()

    # Calculate beta (relative to the specified market index)
    cov_with_market = simple_returns[ticker].cov(simple_returns[market_index])
    beta = cov_with_market / simple_returns[market_index].var()

    # Calculate Sharpe ratio
    sharpe_ratio = (mean_returns[ticker] / std_dev)

    # Store the values in the DataFrame
    risk_measures.loc[ticker] = [std_dev, beta, sharpe_ratio]

# Print risk measures for each asset
print("Risk Measures for Each Asset:")
print(risk_measures)
```

## Limitations of Markowitz optimization and its real-world applications.

Markowitz optimization, a cornerstone of modern portfolio theory, comes with inherent limitations that impact its real-world applications. One major limitation lies in the assumption of normality in asset returns. Financial markets often exhibit non-normal distributions, leading to challenges in accurately modeling extreme events. Estimation error poses another challenge. Markowitz optimization heavily relies on historical data for estimating expected returns and covariances. However, this historical data may not be a reliable predictor of future market conditions. The sensitivity of the model to input parameters is a critical consideration. Small changes in expected returns and covariance matrices can result in significant shifts in portfolio allocations, making the optimization results less robust. Markowitz optimization is a single-period model, disregarding the dynamic nature of financial markets over time. In reality, investment decisions span multiple time periods, requiring models that account for changing risk and return expectations. Transaction costs, a crucial aspect of real-world trading, are not considered in Markowitz optimization. Ignoring these costs can lead to suboptimal portfolio strategies and impractical trading recommendations. The Black-Scholes-Merton framework, consistent with Markowitz assumptions, assumes constant volatility and risk-free rates. However, these parameters can change over time, especially during periods of market stress. The lack of short selling constraints in Markowitz optimization allows for flexibility but introduces complexities, including potential unlimited losses and practical constraints on short-selling. Despite these limitations, Markowitz optimization has played a foundational role in shaping portfolio theory. Real-world applications often involve augmenting the model with risk management techniques and adjustments to better align with the complexities of financial markets.
